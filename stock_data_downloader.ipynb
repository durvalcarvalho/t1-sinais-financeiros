```python
import pandas as pd
import yfinance as yf
from retry import retry
from time import sleep


def get_tickets(df):
    for ticket_orig in df.ticket:
        for sufix in ['3.SA', '4.SA']:
            ticket = ticket_orig.replace('.SA', sufix)
            yield ticket

@retry(Exception, tries=3, delay=1)
def get_historical_prices(stock_name):
    return yf.Ticker(stock_name).history(period='max')


df = pd.read_csv('tickets.csv')
df
```

```python
tickets = list(get_tickets(df))
size = len(tickets)

for i, ticket in enumerate(tickets):
    df_ = get_historical_prices(ticket)
    ticket = ticket.replace('.SA', '')

    if len(df_) > 0:
        df_.to_csv(f'stocks/{ticket}.csv', index=True)
    
    perc = (i+1) / size
    print(f'[{perc:.2%}] - {i+1}/{size}')
    sleep(0.5)
```
